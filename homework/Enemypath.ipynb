Just trying to figure out how to calculate speed B...

An enemy can move one pixel per X tics, but it alternates between speed A and speed B to decide how far to move.

```python
import matplotlib.pyplot as plt
import math
from ipywidgets import interactive

class Speed:
    def __init__(self):
        self.reverse_direction = False
        self.one_pixel_per_x_tics = False
        self.use_speed_b = False
        self.speed_b = 0
        self.speed_a = 0

    def move_amount(self, lifetime):
        # Ticks-per-pixel movement
        speed = self.speed_b if self.use_speed_b else self.speed_a
        if lifetime == speed:
            self.use_speed_b = self.speed_b != 0 and not self.use_speed_b
            return 1 * (2 * bool(self.reverse_direction) - 1)
        return 0

friend_x = 222
friend_y = 72
enemy_x = -15
enemy_speed_x = 1
run = friend_x - enemy_x

def plot_enemy_path(enemy_y=80, speed_b_override=0):
    rise = enemy_y - friend_y
    slope = 0 if rise == 0 else abs(run / rise)
    slope_real, _ = math.modf(slope)
    speed_y = Speed()
    speed_y.speed_a = min(255.0, math.floor(slope))
    #speed_y.speed_b = min(31.0, 0 if slope_real == 0 else math.floor(1./slope_real));
    #speed_y.speed_b = min(255, math.floor(slope * (1 - slope_real if slope_real < .5 else slope_real)))
    speed_y.speed_b = min(255, math.floor(slope * (0.5 + 0.5 * slope_real)))
    speed_y.speed_b = speed_b_override if speed_b_override else speed_y.speed_b
    print("slope", slope + slope_real)
    # speed_y.speed_b = 31
    speed_y.one_pixel_per_x_tics = True
    speed_y.reverse_direction = (rise < 0);
    enemy_yy = enemy_y
    enemy_lifetime = 0
    x_sim_data = []
    y_sim_data = []
    lifetime_y = []
    y_mov_data = []
    for enemy_xx in range(enemy_x, friend_x, enemy_speed_x):
        x_sim_data.append(enemy_xx)
        y_move = speed_y.move_amount(enemy_lifetime)
        y_mov_data.append(y_move)
        enemy_lifetime = (y_move == 0) * (enemy_lifetime + 1)
        lifetime_y.append(enemy_lifetime)
        enemy_yy += y_move
        y_sim_data.append(enemy_yy)
    plt.plot([enemy_x, friend_x], [enemy_y, friend_y])
    plt.plot(x_sim_data, y_sim_data)
    # plt.hist(lifetime_y)
    plt.hist(y_mov_data)
    plt.axis(xmin=0,xmax=240,ymin=0,ymax=160)
    plt.ylabel('Enemy Y')
    plt.show()

w=interactive(plot_enemy_path,enemy_y=(8, 136),speed_b_override=(0,48))
w
```
